```python
from faker import Faker
import pandas as pd
import random, json

fake = Faker()

with open("./mock-insert/studiengaenge.json") as file:
    studiengaenge = json.load(file)

referenten = []
studenten = []

for _ in range(30):
    name = fake.name().split(" ")

    referenten.append({
        "firstName": name[0],
        "lastName": name[1],
        "address": fake.address().replace("\n", " "),
        "email": name[0].lower() + "." + name[1].lower() + "@zdh.thm.de",
        "phoneNumber": f'+49 {fake.msisdn()[3:]}',
        "academicLevel": random.choice(["NONE", "BACHELOR", "MASTER", "DR", "PROF", "PROF_DOCTOR", "DIPLOMA"]),
        "role": random.choice(["PROFESSOR", "EXTERNAL"])
    })

for _ in range(50):
    name = fake.name().split(" ")
    
    studenten.append({
        "firstName": name[0],
        "lastName": name[1],
        "address": fake.address().replace("\n", " "),
        "email": name[0].lower() + "." + name[1].lower() + "@zdh.thm.de",
        "phoneNumber": f'+49 {fake.msisdn()[3:]}',
        "studentNumber": random.randint(480293, 5930404),
        "salutation": random.choice(["HERR", "FRAU", "DIVERS"]),
        "academicLevel": random.choice(["NONE", "BACHELOR", "MASTER", "DR", "DIPLOMA"]),
    })


pd.DataFrame(referenten).to_json("./mock-insert/referenten.json", orient="records")
pd.DataFrame(studenten).to_json("./mock-insert/studenten.json", orient="records")
```
